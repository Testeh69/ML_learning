```python
import tensorflow as tf
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from tensorflow import keras
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
assert x_train.shape == (60000, 28, 28)
assert x_test.shape == (10000, 28, 28)
assert y_train.shape == (60000,)
assert y_test.shape == (10000,)
```

```python
x_train_new = x_train[:]/255.0
x_test_new = x_test[:]/255.0
```

```python
from tensorflow.keras.utils import to_categorical

# Nombre de classes dans votre problème de classification
num_classes = 10  # Par exemple, s'il y a 10 classes différentes

y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
```

```python
learning_rate = 0.00005
best_learning_rate = 0
record_accuracy = 0
while learning_rate < 10:
  print("lerning_rate_actuel: ", learning_rate)
  model = Sequential()
  model.add(Flatten(input_shape=(28,28)))
  model.add(Dense(32, activation = "relu"))
  model.add(Dense(32, activation = "relu"))
  model.add(Dense(32, activation = "relu"))
  model.add(Dense(32, activation = "relu"))
  model.add(Dense(10, activation = "softmax"))
  model.summary()
  model.compile(loss="categorical_crossentropy", optimizer = keras.optimizers.Adam(learning_rate=learning_rate), metrics = ["categorical_accuracy"])
  history = model.fit(x = x_train_new, y = y_train, batch_size = 32, epochs = 300, verbose = 2, validation_split= 0.2, callbacks= keras.callbacks.EarlyStopping(monitor = "val_categorical_accuracy", patience = 10, restore_best_weights=True))
  if history.history["val_categorical_accuracy"][-1] > record_accuracy:
    record_accuracy =  history.history["val_categorical_accuracy"][-1]
    best_learning_rate = learning_rate
  learning_rate = learning_rate * np.exp(np.log(pow(10,6))/500)
```

```python
model.evaluate(x_test_new, y_test)
```
