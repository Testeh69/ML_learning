```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import BatchNormalization
```

```python
X_train.shape
```

```python
model_nn = Sequential()
model_nn.add(BatchNormalization(input_shape = (8,)))
model_nn.add(Dense(256, activation ="relu", input_shape = (8,)))
model_nn.add(BatchNormalization())
model_nn.add(Dense(1))
model_nn.summary()
```

```python
model_nn.compile(optimizer = Adam(), loss = "mae", metrics = ["mae"] )
history = model_nn.fit(X_train,y_train, epochs = 100, validation_data = (X_test,y_test), verbose = 2)
print(history.history.keys())
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.legend(["train", "test"], loc="upper left")
plt.title("model loss")
plt.show()
plt.plot(history.history["mae"])
plt.plot(history.history["val_mae"])
plt.legend(["train", "test"], loc="upper right")
plt.title("mae")
plt.show()
loss, mae = model_nn.evaluate(X_test,y_test)
print(loss,mae)
```

```python
predict = model_nn.predict(X_train.values)
print("result ->:",predict[0],"predictions->",y_train.values[0])
```
